```python
import tc

plt.rc('font', **{'family': 'serif', 'serif': ['Computer Modern Roman']})
plt.rc('text', usetex=True)
fontsize = 16
plt.rc('axes', labelsize=fontsize)
plt.rc('xtick', labelsize=fontsize)
plt.rc('ytick', labelsize=fontsize)
plt.rc('legend', fontsize=fontsize)

# Reward curve figure.
x = [0, 1, 6, 10]
e = [1, .7, .2, .05]
l = [1, 0, 0, 0]

fig = plt.figure(figsize=(8, 5), dpi=300)
ax = fig.add_subplot(111)
ax.plot(x, e, '--sk', label='Entropy', markersize=8, linewidth=3)
ax.plot(x, l, '--or', label='Loss', markersize=8, linewidth=3)
ax.vlines(10, -1, 2, 'b', linewidth=4)
# ax.vlines(1, .7, -1, 'k', linewidth=3)

cm = mpl.cm.Set2

# This is a hack to properly render hatching to pdf
# from http://stackoverflow.com/questions/5195466/
ax.fill_betweenx([1, .7], [0, 1], [10, 10], color=cm(0.0), alpha=0.15)
ax.fill_betweenx([1, .7], [0, 1], [10, 10], color='None', alpha=0.15, edgecolor='k', hatch="/")

ax.fill_betweenx([.7, .2], [1, 6], [10, 10], color=cm(0.5), alpha=0.15)
# ax.fill_betweenx([.7, .2], [1, 6], [10, 10], color='None', alpha=0.15, edgecolor='k', hatch='.')

ax.fill_betweenx([.2, .05], [6, 10], [10, 10], color=cm(1.0), alpha=0.15, hatch="|")
ax.fill_betweenx([.2, .05], [6, 10], [10, 10], color='None', alpha=0.15, edgecolor='k', hatch="|")

xlim([0, 10.5])
xlabel('Cumulative cost')
ylim([0, 1])

spines_to_remove = ['top', 'right']
for spine in spines_to_remove:
    ax.spines[spine].set_visible(False)
    
ax.xaxis.set_ticks_position('none')
ax.yaxis.set_ticks_position('none')

almost_black = '#262626'
spines_to_keep = ['bottom', 'left']
for spine in spines_to_keep:
    ax.spines[spine].set_linewidth(0.5)
    
legend()
plt.savefig('rewards_src.pdf', facecolor='none')
```
