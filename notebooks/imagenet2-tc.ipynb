```python
%load_ext autoreload
%autoreload 2
from scipy.io import loadmat
import tc
from tc.hedging import *

data_sources_dir = '../data/data_sources'
imagenet = tc.data_sources.imagenet.ILSVRC65(data_sources_dir)

num_guarantees = 5  # TODO increase for more fine-grained (and slower) evaluation
s = 2  # controls the fine-grainedness at the top accuracies
accuracy_guarantees = np.array(sorted((10 ** s - logspace(0, s, num_guarantees)) / 10 ** s))
accuracy_guarantees = accuracy_guarantees / accuracy_guarantees.max() * .99
accuracy_guarantees
```

```python
budgets = range(1, confidences[0].shape[0])

budget_confs = [np.array([_[i] for _ in confidences]) for i in budgets]
print len(budget_confs)
print budget_confs[0].shape
```

```python
train_budget_confs = [np.array([_[i] for _ in train_confidences]) for i in budgets]
print len(train_budget_confs)
print train_budget_confs[0].shape
```

```python
# We can see how the confidences change with the budget
matshow(budget_confs[0][:20])
matshow(budget_confs[5][:20])
matshow(budget_confs[20][:20])
matshow(budget_confs[-1][:20])
```

```python
# Fill blanks with prior
# Replace 0 readings with mean confidences on the training set

prior = np.tile(imagenet.X.mean(0), (imagenet.X_test.shape[0], 1))
for budget_conf in budget_confs:
    budget_conf[budget_conf == 0] = prior[budget_conf == 0]
```

```python
train_prior = np.tile(imagenet.X.mean(0), (imagenet.X.shape[0], 1))
for budget_conf in train_budget_confs:
    budget_conf[budget_conf == 0] = train_prior[budget_conf == 0]
```

```python
lambdas_method='empirical'

confidence = .95
num_bs_iters = 30
if lambdas_method == 'original':
    lambdas = darts_bisection(imagenet.X, accuracy_guarantees,
                              imagenet.y, imagenet.graph, num_bs_iters, confidence)
elif lambdas_method == 'empirical':
    # TODO: introduce sampling here?
    lambdas = darts_bisection(np.vstack(train_budget_confs), accuracy_guarantees,
                              np.tile(imagenet.y, (len(budgets), 1)), imagenet.graph, num_bs_iters, confidence)

r, a, hp, ha = zip(*[
    darts_eval(budget_conf, imagenet.y_test, lambdas, imagenet.graph)
    for budget_conf in budget_confs])
rewards = np.array(r)
accuracies = np.array(a)
height_portions = np.dstack(hp)
height_accs = np.dstack(ha)

# add the rewards = 0, accuracies = 1 point to all curves
rewards = np.hstack((rewards, np.zeros((rewards.shape[0], 1))))
accuracies = np.hstack((accuracies, np.ones((accuracies.shape[0], 1))))
```

```python
graph = imagenet.graph

fig = plot_accuracy_vs_specificity(
    rewards, accuracies, labels=['{:.1f}'.format(b) for b in budgets])
```

```python
print lambdas.shape

# the shape is (num possible heights, lambdas, budgets)
print height_portions.shape

# for acc=0.9
cmap = plt.cm.winter
for h in range(height_portions.shape[0]):
    plot(height_portions[h, 2, :], label='height {}'.format(h), color=cmap(1. * h / height_portions.shape[0]))
legend()

## 
g = graph['g']
nodes = graph['nodes']
root = [n for n in g.nodes() if g.predecessors(n) == []]
pos = nx.pygraphviz_layout(g, prog='twopi')
# pos = nx.graphviz_layout(g, prog='circo', root=root)
# pos = nx.spring_layout(g)
# pos=nx.circular_layout(g)
pos=nx.spring_layout(g,dim=2,pos=pos, iterations=2, fixed=root)

gray = (0.75, 0.75, 0.75)
graydark = (0.5, 0.5, 0.5)
ns = 120

labels = dict(zip(nodes, [g.node[node]['word'].split(',')[0] for node in nodes]))
leaf_nodes = np.array(graph['nodes'])[graph['heights'] == 0]
for node in leaf_nodes:
    labels[node] = ''

plt.rc('font', **{'family': 'serif', 'serif': ['Computer Modern Roman']})
plt.rc('text', usetex=True)
fontsize=16
plt.rc('axes', labelsize=fontsize)
plt.rc('xtick', labelsize=fontsize)
plt.rc('ytick', labelsize=fontsize)
plt.rc('legend', fontsize=fontsize)
    
# shorten name
for k in labels.keys():
    if labels[k] == 'domestic cat':
        labels[k] = 'cat'
    
cmap = plt.get_cmap('Reds')
fig = plt.figure(figsize=(20, 5.5))
num_budget_points = 4

from mpl_toolkits.axes_grid1 import Grid
for i, b in enumerate(np.linspace(0, len(budgets)-1, num_budget_points).astype('int')):
    ax = fig.add_subplot(1, num_budget_points, i+1)
    for h in np.unique(graph['heights']):
        h_nodes = np.array(graph['nodes'])[graph['heights'] == h].tolist()
        cval = np.minimum(1, height_portions[h, accuracy_guarantee_ind, b])
        cval *= 1.5  # blow the color differences out a bit
        drawn_nodes = nx.draw_networkx_nodes(
            g, pos, nodelist=h_nodes, node_size=ns, node_color=[cval] * len(h_nodes),
            ax=ax, cmap=cmap, vmin=0, vmax=1)
        drawn_nodes.set_edgecolor(graydark)
        
        # Node labels
        if h > 0:
            pos_offset = pos.copy()
            for k in pos_offset.keys():
                pos_offset[k] = (pos_offset[k][0] - 0.025, pos_offset[k][1] - 0.0525)
                    
            nx.draw_networkx_labels(
                g, pos_offset, labels,
                font_family='serif', font_size=fontsize, font_color='gray',
                horizontalalignment='left')
    
    nx.draw_networkx_edges(g, pos, arrows=False, edge_color=[gray] * len(g.edges()), ax=ax)
    ax.text(0.3, -.15, 'Cost: {}'.format(budgets[b]), fontsize=20)
    plt.axis('off')
    plt.axis('equal')
    
fig.subplots_adjust(right=0.92)
# [left, bottom, width, height]
cbar_ax = fig.add_axes([0.95, 0.2, 0.01, 0.6])
plt.colorbar(drawn_nodes, cax=cbar_ax)

# fig.tight_layout()
plt.savefig('../writeup/figures/apr11_assembly/ilsvrc65_network.pdf')
```
