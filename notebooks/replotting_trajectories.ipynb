```python
base = 'data/timely_results/synthetic_orthants_D3_False_N16000_Nt4000_13'
experiments = [
    #'linear_untaken-dynamic-1.0-True-logreg-3-False-0-auc-infogain-False-20-0.2-10',
    #'linear_untaken-static-1.0-True-logreg-1-False-0-auc-infogain-False-10-0.2-5',
    #'manual_orthants-None-0.0-False-logreg-1-False-0-auc-infogain-False-2-1.0-8',
    #'random-None-0.0-True-logreg-1-False-0-auc-infogain-False-4-1.0-8',
    'linear_untaken-dynamic-0.0-True-logreg-1-False-0-auc-infogain-False-10-0.2-5'
]
output_dirname = assembly_dirname + '/' + os.path.basename(base)
if not os.path.exists(output_dirname):
    os.makedirs(output_dirname)

for experiment in experiments:
    print experiment
    dirname = base + '/' + experiment
    
    # Cache the old pickle file
    pickle_filename = dirname + '/ticl.pickle'
    old_pickle_filename = dirname + '/ticl_old.pickle'
    shutil.copy(pickle_filename, old_pickle_filename)
    
    # Load the classifier from pickle, and re-run evaluation
    timclf = cPickle.load(open(dirname + '/ticl.pickle'))
    timclf.ds.data_filename = timclf.ds.data_filename.replace('/n/banquet/df/sergeyk', '/Users/sergeyk')
    timclf.evaluate(num_workers=4, force=True)
    
    # Copy the plot
    shutil.copy(dirname + '/trajectories_final.png.pdf', output_dirname + '/{}.pdf'.format(experiment))
    
    # Restore the old pickle file.
    shutil.move(old_pickle_filename, pickle_filename)
```
